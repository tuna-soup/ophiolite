# Ophiolite Log-to-AVO Golden Path

This notebook is the teachable front door for the LAS/log to Zoeppritz AVO workflow.

It stays on the Python SDK side of the product boundary:

- ingest and canonical data/compute happen in `ophiolite`
- chart rendering stays downstream in `Ophiolite Charts`


```python
from pathlib import Path
import json
import sys

def find_repo_root(start: Path) -> Path:
    for candidate in (start, *start.parents):
        if (candidate / "Cargo.toml").exists() and (candidate / "python" / "src").exists():
            return candidate
    raise RuntimeError("could not locate the Ophiolite repo root from the current working directory")

REPO_ROOT = find_repo_root(Path.cwd())
NOTEBOOK_DIR = REPO_ROOT / "examples" / "golden_paths" / "log_avo"
if str(NOTEBOOK_DIR) not in sys.path:
    sys.path.insert(0, str(NOTEBOOK_DIR))

from log_avo_workflow import detect_real_las_paths, run_workflow

REPO_ROOT
```

## Data selection

If the real F3 F02-A-05 LAS pair is available, the workflow uses it. Otherwise it falls back to the synthetic project fixture.

```python
dsi_path, density_path = detect_real_las_paths()
data_mode = "real" if dsi_path is not None and density_path is not None else "synthetic"
{
    "data_mode": data_mode,
    "dsi_las_path": None if dsi_path is None else str(dsi_path),
    "density_las_path": None if density_path is None else str(density_path),
}
```

## Run the workflow

This creates a project, ingests the logs, resolves the elastic channels, builds AVO interfaces, and writes chart-ready DTO JSON under a temporary run root.

```python
summary = run_workflow(
    data_mode=data_mode,
    interval_thickness_m=6.096,
    materialize_derived_channels=False,
)
summary
```
